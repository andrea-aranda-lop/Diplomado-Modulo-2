# Probabilidad

## 1. Lenguaje básico de conjuntos para probabilidad

En probabilidad, los resultados posibles de un experimento se representan con conjuntos.

| Símbolo | Nombre | Significado |
|---|---|---|
| $\Omega$ | Espacio muestral | Conjunto de todos los resultados posibles |
| $A, B$ | Eventos | Subconjuntos de $\Omega$ |
| $A \cup B$ | Unión | Ocurre $A$ o ocurre $B$ o ambos |
| $A \cap B$ | Intersección | Ocurren $A$ y $B$ al mismo tiempo |
| $A^c$ | Complemento | No ocurre $A$ |
| $\varnothing$ | Conjunto vacío | Evento imposible |
| $A \subseteq B$ | Subconjunto | Si ocurre $A$, entonces también ocurre $B$ |

### Ejemplo

Si lanzamos un dado:

$$\Omega = \{1,2,3,4,5,6\}$$

Podemos definir:

- $A =$ “sale número par” $= \{2,4,6\}$
- $B =$ “sale número mayor que 3” $= \{4,5,6\}$

Entonces:

$$A \cap B = \{4,6\}$$

$$A \cup B = \{2,4,5,6\}$$

$$A^c = \{1,3,5\}$$

```python
# Ejemplo con conjuntos en Python
Omega = {1, 2, 3, 4, 5, 6}
A = {2, 4, 6}        # número par
B = {4, 5, 6}        # número mayor que 3

print("Omega:", Omega)
print("A unión B:", A | B)
print("A intersección B:", A & B)
print("Complemento de A:", Omega - A)
```

## 2. Experimento aleatorio y evento aleatorio

### Experimento aleatorio

Un **experimento aleatorio** es un proceso cuyo resultado no se puede predecir con certeza antes de realizarlo, aunque sí se conocen los posibles resultados.

Ejemplos:

- Lanzar una moneda.
- Lanzar un dado.
- Elegir una persona al azar de una población.
- Medir el tiempo de espera de un cliente.
- Revisar si una transacción fue aprobada o rechazada.

### Evento aleatorio

Un **evento aleatorio** es un conjunto de resultados posibles del experimento.

Formalmente, si $\Omega$ es el espacio muestral, un evento $A$ es un subconjunto de $\Omega$:

$$A \subseteq \Omega$$

### Ejemplo

Experimento: lanzar un dado.

Espacio muestral:

$$\Omega = \{1,2,3,4,5,6\}$$

Evento:

$$A = \text{“sale un número par”} = \{2,4,6\}$$

El evento $A$ ocurre si el resultado observado pertenece a $A$.

## 3. Espacio muestral y sus componentes

El **espacio muestral** es el conjunto de todos los resultados posibles de un experimento aleatorio.

Se suele representar con:

$$\Omega$$


### Componentes principales

| Componente | Descripción |
|---|---|
| Resultado elemental | Cada resultado individual posible |
| Espacio muestral $\Omega$ | Conjunto de todos los resultados elementales |
| Evento | Subconjunto del espacio muestral |
| Evento seguro | El propio espacio muestral $\Omega$ |
| Evento imposible | El conjunto vacío $\varnothing$ |

### Ejemplo: lanzamiento de dos monedas

Experimento: lanzar dos monedas.

$$\Omega = \{CC, CS, SC, SS\}$$

Donde:

- $C =$ cara
- $S =$ sello

Evento $A =$ “sale exactamente una cara”:

$$A = \{CS, SC\}$$

Evento $B =$ “sale al menos una cara”:

$$B = \{CC, CS, SC\}$$

```python
# Construcción del espacio muestral para dos monedas
moneda = ["C", "S"]
Omega = [(m1, m2) for m1 in moneda for m2 in moneda]

A = [resultado for resultado in Omega if resultado.count("C") == 1]
B = [resultado for resultado in Omega if resultado.count("C") >= 1]

print("Espacio muestral:", Omega)
print("A = exactamente una cara:", A)
print("B = al menos una cara:", B)
```

## 4. Espacio de probabilidad y sus componentes

Un **espacio de probabilidad** es una estructura matemática que permite asignar probabilidades a eventos.

Se representa como una terna:

$$\left(\Omega, \mathcal{F}, P\right)$$

Donde:

| Componente | Significado |
|---|---|
| $\Omega$ | Espacio muestral: todos los resultados posibles |
| $\mathcal{F}$ | Familia de eventos a los que se puede asignar probabilidad |
| $P$ | Función de probabilidad |

### Interpretación

- $\Omega$ responde: **¿qué puede pasar?**
- $\mathcal{F}$ responde: **¿sobre qué eventos puedo hablar?**
- $P$ responde: **¿qué tan probable es cada evento?**

### Ejemplo con un dado justo

$$\Omega = \{1,2,3,4,5,6\}$$

Evento:

$$A = \{2,4,6\}$$

Como el dado es justo:

$$P(1)=P(2)=P(3)=P(4)=P(5)=P(6)=\frac{1}{6}$$

Entonces:

$$P(A)=P(\{2,4,6\})=\frac{3}{6}=\frac{1}{2}$$

## 5. Axiomas de Kolmogorov

Los axiomas de Kolmogorov son las reglas fundamentales que debe cumplir cualquier función de probabilidad.

Sea $P$ una función que asigna probabilidades a eventos de un espacio muestral $\Omega$.

### Axioma 1: No negatividad

Para cualquier evento $A$:

$$P(A) \geq 0$$

Una probabilidad nunca puede ser negativa.

### Axioma 2: Normalización

La probabilidad del espacio muestral completo es 1:

$$P(\Omega)=1$$

Esto significa que algo dentro de los resultados posibles necesariamente ocurre.

### Axioma 3: Aditividad

Si $A$ y $B$ son eventos mutuamente excluyentes, es decir:

$$A \cap B = \varnothing$$

entonces:

$$P(A \cup B)=P(A)+P(B)$$

Para una colección de eventos mutuamente excluyentes $A_1,A_2,A_3,\dots$:

$$P\left(\bigcup_i A_i\right)=\sum_i P(A_i)$$

### Consecuencias importantes

A partir de los axiomas se obtienen reglas muy usadas:

$$P(A^c)=1-P(A)$$

$$P(\varnothing)=0$$

$$0 \leq P(A) \leq 1$$

Si $A \subseteq B$, entonces:

$$P(A) \leq P(B)$$

## 6. Propiedades de probabilidad

### Regla del complemento

Si $A^c$ representa que **no ocurre $A$**, entonces:

$$P(A^c)=1-P(A)$$

### Regla de la unión para dos eventos

Para dos eventos cualesquiera $A$ y $B$:

$$P(A \cup B)=P(A)+P(B)-P(A \cap B)$$

Se resta $P(A \cap B)$ porque la intersección se cuenta dos veces.

### Eventos mutuamente excluyentes

Dos eventos son **mutuamente excluyentes** si no pueden ocurrir al mismo tiempo:

$$A \cap B = \varnothing$$

En ese caso:

$$P(A \cup B)=P(A)+P(B)$$

### Ejemplo

En una clase:

- $P(A)=0.60$: probabilidad de que un estudiante apruebe Matemáticas.
- $P(B)=0.50$: probabilidad de que apruebe Estadística.
- $P(A \cap B)=0.30$: probabilidad de que apruebe ambas.

Entonces:

$$P(A \cup B)=0.60+0.50-0.30=0.80$$

La probabilidad de que apruebe al menos una de las dos materias es 0.80.

```python
# Regla de la unión
P_A = 0.60
P_B = 0.50
P_A_inter_B = 0.30

P_A_union_B = P_A + P_B - P_A_inter_B
print("P(A unión B) =", P_A_union_B)
```

## 7. Probabilidad clásica

La **probabilidad clásica** se usa cuando todos los resultados del espacio muestral son igualmente probables.

La fórmula es:

$$P(A)=\frac{\text{número de casos favorables}}{\text{número de casos posibles}}$$

### Ejemplo: dado justo

Experimento: lanzar un dado.

Evento $A =$ “sale número par”.

$$A=\{2,4,6\}$$

Casos favorables: 3.

Casos posibles: 6.

$$P(A)=\frac{3}{6}=\frac{1}{2}$$

### Cuándo usarla

Se usa cuando:

- El espacio muestral es finito.
- Los resultados tienen la misma posibilidad de ocurrir.
- Es razonable asumir simetría, como en dados o monedas ideales.

### Cuidado

No conviene usar probabilidad clásica si los resultados no son igualmente probables.

Por ejemplo, si una moneda está cargada, no necesariamente:

$$P(\text{cara})=\frac{1}{2}$$

## 8. Técnicas básicas de conteo

Antes de estudiar variables aleatorias, conviene dominar algunas técnicas de conteo, porque muchas probabilidades clásicas dependen de contar casos favorables y casos posibles.

### Principio multiplicativo

Si una tarea puede hacerse en $m$ formas y otra en $n$ formas, entonces ambas tareas pueden hacerse en:

$$m \times n$$

formas.

Ejemplo: si tienes 3 camisas y 2 pantalones, puedes formar:

$$3 \times 2 = 6$$

combinaciones.

### Permutaciones

Se usan cuando **el orden importa**.

Número de formas de ordenar $n$ objetos distintos:

$$n!$$

Número de formas de ordenar $r$ objetos tomados de $n$:

$$P(n,r)=\frac{n!}{(n-r)!}$$

### Combinaciones

Se usan cuando **el orden no importa**.

Número de formas de elegir $r$ objetos de un total de $n$:

$$C(n,r)=\binom{n}{r}=\frac{n!}{r!(n-r)!}$$

### Ejemplo

De 5 personas se eligen 2 para formar un comité.

Como el orden no importa:

$$\binom{5}{2}=\frac{5!}{2!3!}=10$$

```python
import math

n = 5
r = 2

permutaciones = math.factorial(n) // math.factorial(n - r)
combinaciones = math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

print(f"Permutaciones P({n},{r}) =", permutaciones)
print(f"Combinaciones C({n},{r}) =", combinaciones)
```

## 9. Probabilidad geométrica

La **probabilidad geométrica** se usa cuando los resultados posibles forman una región continua, como una línea, un área o un volumen.

La idea es comparar medidas geométricas.

### En una dimensión

$$P(A)=\frac{\text{longitud favorable}}{\text{longitud total}}$$

### En dos dimensiones

$$P(A)=\frac{\text{área favorable}}{\text{área total}}$$

### En tres dimensiones

$$P(A)=\frac{\text{volumen favorable}}{\text{volumen total}}$$

### Ejemplo

Se elige un número al azar entre 0 y 10.

Evento $A =$ “el número cae entre 2 y 5”.

Longitud favorable:

$$5-2=3$$

Longitud total:

$$10-0=10$$

Entonces:

$$P(A)=\frac{3}{10}=0.30$$

### Otro ejemplo 

Si lanzas un dardo a una tabla cuadrada y quieres calcular la probabilidad de caer dentro de un círculo dibujado en ella:

$$P(\text{caer en el círculo})=\frac{\text{área del círculo}}{\text{área del cuadrado}}$$

```python
# Simulación de probabilidad geométrica:
# elegir un número aleatorio entre 0 y 10 y ver si cae entre 2 y 5
import random

n_simulaciones = 100_000
favorables = 0

for _ in range(n_simulaciones):
    x = random.uniform(0, 10)
    if 2 <= x <= 5:
        favorables += 1

prob_estimada = favorables / n_simulaciones
print("Probabilidad estimada:", prob_estimada)
print("Probabilidad teórica:", 3 / 10)
```

## 10. Probabilidad frecuentista

La **probabilidad frecuentista** interpreta la probabilidad como la frecuencia relativa con la que ocurre un evento cuando el experimento se repite muchas veces.

Si un experimento se repite $n$ veces y el evento $A$ ocurre $n_A$ veces, entonces:

$$P(A) \approx \frac{n_A}{n}$$

Cuando $n$ crece mucho, la frecuencia relativa tiende a estabilizarse alrededor de la probabilidad real.

### Ejemplo

Si lanzamos una moneda 1,000 veces y obtenemos cara 520 veces:

$$P(\text{cara}) \approx \frac{520}{1000}=0.52$$

### Diferencia contra probabilidad clásica

| Enfoque | Idea principal | Ejemplo |
|---|---|---|
| Clásica | Casos favorables / casos posibles | Dado justo |
| Geométrica | Medida favorable / medida total | Área dentro de una figura |
| Frecuentista | Frecuencia observada en repeticiones | Simular lanzamientos de moneda |

```python
# Simulación frecuentista: lanzamiento de una moneda
import random
import matplotlib.pyplot as plt

n_lanzamientos = 1_000
caras = 0
frecuencias = []

for i in range(1, n_lanzamientos + 1):
    resultado = random.choice(["C", "S"])
    if resultado == "C":
        caras += 1
    frecuencias.append(caras / i)

plt.figure(figsize=(8, 4))
plt.plot(range(1, n_lanzamientos + 1), frecuencias)
plt.axhline(0.5, linestyle="--")
plt.xlabel("Número de lanzamientos")
plt.ylabel("Frecuencia relativa de cara")
plt.title("Frecuencia relativa en lanzamientos de moneda")
plt.show()
```

## 11. Probabilidad condicional

La **probabilidad condicional** mide la probabilidad de que ocurra un evento $A$, sabiendo que ya ocurrió otro evento $B$.

Se escribe:

$$P(A \mid B)$$

Se lee: “probabilidad de $A$ dado $B$”.

La fórmula es:

$$P(A \mid B)=\frac{P(A \cap B)}{P(B)}$$

siempre que:

$$P(B)>0$$

### Interpretación

Cuando sabemos que ocurrió $B$, el espacio muestral se reduce a $B$.

Dentro de ese nuevo espacio, buscamos qué parte también pertenece a $A$.

### Ejemplo

Lanzamos un dado.

- $A =$ “sale número par” $= \{2,4,6\}$
- $B =$ “sale número mayor que 3” $= \{4,5,6\}$

Entonces:

$$A \cap B = \{4,6\}$$

$$P(A \mid B)=\frac{P(A \cap B)}{P(B)}=\frac{2/6}{3/6}=\frac{2}{3}$$

Sabiendo que salió un número mayor que 3, la probabilidad de que sea par es $2/3$.

```python
# Probabilidad condicional con conjuntos
Omega = {1, 2, 3, 4, 5, 6}
A = {2, 4, 6}  # par
B = {4, 5, 6}  # mayor que 3

P_A_inter_B = len(A & B) / len(Omega)
P_B = len(B) / len(Omega)
P_A_dado_B = P_A_inter_B / P_B

print("P(A ∩ B):", P_A_inter_B)
print("P(B):", P_B)
print("P(A | B):", P_A_dado_B)
```

## 12. Regla del producto

A partir de la probabilidad condicional:

$$P(A \mid B)=\frac{P(A \cap B)}{P(B)}$$

podemos despejar la intersección:

$$P(A \cap B)=P(A \mid B)P(B)$$

También se puede escribir como:

$$P(A \cap B)=P(B \mid A)P(A)$$

### Interpretación

Para que ocurran $A$ y $B$ al mismo tiempo:

1. Primero ocurre uno de los eventos.
2. Después ocurre el otro bajo la condición de que el primero ya ocurrió.

### Ejemplo

En una empresa:

- $P(B)=0.40$: probabilidad de que una persona sea del área de datos.
- $P(A \mid B)=0.25$: probabilidad de que una persona use Python dado que es del área de datos.

Entonces:

$$P(A \cap B)=0.25 \times 0.40=0.10$$

La probabilidad de que una persona sea del área de datos y use Python es 0.10.

## 13. Probabilidad total

La **probabilidad total** permite calcular la probabilidad de un evento usando casos o grupos que dividen el espacio muestral.

Supongamos que $B_1, B_2, \dots, B_n$ forman una partición de $\Omega$.

Eso significa que:

1. Cubren todo el espacio muestral.
2. No se traslapan entre sí.

Entonces:

$$P(A)=P(A \mid B_1)P(B_1)+P(A \mid B_2)P(B_2)+\cdots+P(A \mid B_n)P(B_n)$$

De forma compacta:

$$P(A)=\sum_{i=1}^{n} P(A \mid B_i)P(B_i)$$

### Ejemplo

Una empresa recibe archivos de dos fuentes:

- Fuente 1 produce el 70% de los archivos: $P(F_1)=0.70$
- Fuente 2 produce el 30% de los archivos: $P(F_2)=0.30$

Probabilidad de error por fuente:

- $P(E \mid F_1)=0.02$
- $P(E \mid F_2)=0.05$

Entonces:

$$P(E)=0.02(0.70)+0.05(0.30)=0.014+0.015=0.029$$

La probabilidad total de recibir un archivo con error es 0.029, es decir, 2.9%.

```python
# Probabilidad total
P_F1 = 0.70
P_F2 = 0.30
P_E_dado_F1 = 0.02
P_E_dado_F2 = 0.05

P_E = P_E_dado_F1 * P_F1 + P_E_dado_F2 * P_F2
print("P(E) =", P_E)
```

## 14. Teorema de Bayes

El **teorema de Bayes** permite actualizar una probabilidad cuando se conoce nueva evidencia.

La fórmula básica es:

$$P(A \mid B)=\frac{P(B \mid A)P(A)}{P(B)}$$

Donde:

| Término | Interpretación |
|---|---|
| $P(A)$ | Probabilidad inicial o previa |
| $P(B \mid A)$ | Probabilidad de observar la evidencia $B$ si $A$ es cierto |
| $P(B)$ | Probabilidad total de la evidencia |
| $P(A \mid B)$ | Probabilidad actualizada después de observar $B$ |

### Ejemplo usando el caso de archivos con error

Queremos saber la probabilidad de que un archivo venga de la Fuente 2 dado que tiene error:

$$P(F_2 \mid E)=\frac{P(E \mid F_2)P(F_2)}{P(E)}$$

Sabemos que:

- $P(E \mid F_2)=0.05$
- $P(F_2)=0.30$
- $P(E)=0.029$

Entonces:

$$P(F_2 \mid E)=\frac{0.05(0.30)}{0.029}=0.5172$$

Si un archivo tiene error, la probabilidad de que venga de la Fuente 2 es aproximadamente 51.72%.

```python
# Teorema de Bayes
P_F2_dado_E = (P_E_dado_F2 * P_F2) / P_E
print("P(F2 | E) =", P_F2_dado_E)
```

## 15. Independencia de eventos

Dos eventos $A$ y $B$ son **independientes** si la ocurrencia de uno no cambia la probabilidad del otro.

Formalmente:

$$P(A \mid B)=P(A)$$

siempre que $P(B)>0$.

También se cumple que:

$$P(B \mid A)=P(B)$$

Y una forma muy usada para verificar independencia es:

$$P(A \cap B)=P(A)P(B)$$

### Ejemplo de eventos independientes

Experimento: lanzar una moneda y un dado.

- $A =$ “la moneda cae en cara”
- $B =$ “el dado cae en 6”

El resultado de la moneda no afecta al resultado del dado.

Por tanto:

$$P(A \cap B)=P(A)P(B)=\frac{1}{2}\cdot\frac{1}{6}=\frac{1}{12}$$

### Ejemplo de eventos no independientes

Tomar una carta de una baraja sin reemplazo.

- $A =$ “la primera carta es as”
- $B =$ “la segunda carta es as”

Si la primera carta fue as, quedan menos ases disponibles para la segunda carta. Entonces los eventos no son independientes.

### Importante: independiente no significa mutuamente excluyente

- **Mutuamente excluyentes:** no pueden ocurrir al mismo tiempo.
- **Independientes:** la ocurrencia de uno no cambia la probabilidad del otro.

Si dos eventos con probabilidad positiva son mutuamente excluyentes, entonces normalmente **no son independientes**, porque si ocurre uno, el otro ya no puede ocurrir.

```python
# Verificación de independencia con valores de probabilidad
P_A = 0.5       # moneda cae en cara
P_B = 1 / 6     # dado cae en 6
P_A_inter_B = 1 / 12

son_independientes = abs(P_A_inter_B - (P_A * P_B)) < 1e-12
print("¿Son independientes?", son_independientes)
```

## 16. Independencia de más de dos eventos

Para tres eventos $A$, $B$ y $C$, no basta con que sean independientes por pares.

Para que sean mutuamente independientes, deben cumplirse todas estas condiciones:

$$P(A \cap B)=P(A)P(B)$$

$$P(A \cap C)=P(A)P(C)$$

$$P(B \cap C)=P(B)P(C)$$

Y además:

$$P(A \cap B \cap C)=P(A)P(B)P(C)$$

### Idea clave

La independencia por pares no garantiza independencia conjunta.

Este detalle se vuelve importante cuando se estudian variables aleatorias, modelos probabilísticos y distribuciones.
